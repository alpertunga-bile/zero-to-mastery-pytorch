```python
torch.manual_seed(0)
randomTensor = torch.rand(size=(7, 7))
randomTensor
```

```python
torch.manual_seed(0)
randomTensorA = torch.rand(size=(1, 7))
torch.manual_seed(0)
randomTensorB = torch.rand(size=(1, 7))
randomTensorA.matmul(randomTensorB.T)
```

```python
torch.manual_seed(1234)
cpuTensorA = torch.rand(size=(2, 3))
torch.manual_seed(1234)
cpuTensorB = torch.rand(size=(2, 3))

gpuTensorA = cpuTensorA.to("cuda")
gpuTensorB = cpuTensorB.to("cuda")

gpuMatMul = gpuTensorA.matmul(gpuTensorB.T)
```

```python
torch.max(gpuMatMul), torch.min(gpuMatMul)
```

```python
torch.argmax(gpuMatMul), torch.argmin(gpuMatMul)
```
